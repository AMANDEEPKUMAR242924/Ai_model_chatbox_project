```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Model aur Tokenizer ka naam
model_name = "EleutherAI/gpt-neo-2.7B"

print("Model aur Tokenizer load ho rahe hain... Please wait")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

print("Model ready hai! Ab aap chat kar sakte ho. Exit karne ke liye 'exit' type karo.")

# Chat loop
while True:
    user_input = input("\nYou: ")
    
    if user_input.lower() == "exit":
        print("Chat session khatam hua.")
        break

    # Input ko encode karo
    inputs = tokenizer(user_input, return_tensors="pt").to("cuda")  # Use "cpu" if no GPU

    # Model se response generate karo
    outputs = model.generate(**inputs, max_new_tokens=150)

    # Output decode karo
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    print("\nAI:", response)
```
